### Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
data =pd.read_csv("/content/Iris.csv",index_col=0)
```

```python
data.head()
```

```python
#Changes does not effect the "data" dataframe
data1=data.copy()
```

```python
labelencoder = LabelEncoder()
data1.iloc[:, -1] = labelencoder.fit_transform(data1.iloc[:,-1])
```

```python
data1
```

```python
data1['Species'].value_counts()
```

### One Hot Encoder

#### Using sklearn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
data2=pd.read_csv("Iris.csv")
```

```python
data2
```

```python
# creating instance of one-hot-encoder
enc = OneHotEncoder()
```

```python
# passing bridge-types-cat column (label encoded values of bridge_types)
enc_df = pd.DataFrame(enc.fit_transform(data2[['Species']]).toarray())
```

```python
enc_df
```

```python
enc.fit_transform(data2[['Species']])
```

```python
enc_df
```

```python
# merge with main df
data_final = data2.iloc[:,0:5].join(enc_df)
data_final
```
